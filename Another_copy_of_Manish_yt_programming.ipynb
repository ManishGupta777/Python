```python
# create student class that takes name & marks of 3 subjects as arguments
#in constructor. then creat a method to print the avg
class Students:
  def __init__(self,name, marks):
    self.name=name
    self.marks=marks
  def get_avg(self):
    sum=0
    for i in self.marks:
      sum+=i
    print(f"hi {self.name} your avg score is {sum/3}")

s1=Students("manish",[98,97,99])
s2=Students("sam",[90,89,100])
s3=Students("jaggu",[100,98,100])
print(s1.get_avg())
print(s2.get_avg())
print(s3.get_avg())
```

```python
#create account class with 2 attributes balance & account no , create methods for debit,credit& printing the balance
class Account:
  def __init__(self,bal,acc):
    self.balance=bal
    self.account=acc
  def debited(self,amount):
    self.balance-=amount
    print(f"your balance is {self.balance}")
  def credited(self,amount):
    self.balance+=amount
    print(f"your balance is {self.balance}")
acc1=Account(1000,9876543)
acc1.debited(500)
acc1.credited(10000)
```
