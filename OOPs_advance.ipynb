```python
class Atm:
  def __init__(self):
    self.pin=''
    self.balance=0
    self.menu()
  def menu(self):
    user_input=input(""" Hi how can help you
    1. press 1 to create pin
    2. press 2 to change pin
    3. press 3 to chech balance
    4. press 4 to withdraw
    5.anything else to exit """)
    if user_input=='1':
      self.create_pin()
    elif user_input=='2':
      self.pin_update()
    elif user_input=='3':
      self.check_balance()
    elif user_input=='4':
      self.withdraw()
    else:
       exit()
  def create_pin(self):
    user_pin=input("enter yuor pin")
    self.pin=user_pin

    user_balance=int(input("enter bbalance"))
    self.balance=user_balance
    print("pin created successfully")
    self.menu()
  def pin_update(self):
    old_pin=input("enter old pin") # input returns string, comparing with string self.pin
    if(old_pin==self.pin):
      new_pin=input("enter new pin")
      self.pin=new_pin
      print("pin updated successfully")
      self.menu()
    else:
      print("Incorrect pin")
      self.menu()
  def check_balance(self):
    user_input=input("enter your pin") # input returns string, comparing with string self.pin
    if user_input==self.pin:
      print(self.balance)
      self.menu()
    else:
      print("Incorrect pin")
      self.menu()
  def withdraw(self):
    user_input=input("enter your pin") # input returns string, comparing with string self.pin
    if user_input==self.pin:
      amount=int(input("enter your amount"))
      if amount <= self.balance:
        self.balance=self.balance - amount
        print("withdrawn successfully")
        self.menu()
      else:
        print("Insufficient balance. bhag jaa")
        self.menu()
    else:
      print("Incorrect pin. sale chor")
      self.menu()
```

```python
#fraction data type making
class fraction:
  def __init__(self, x, y): #parameteretied constructor hai (required input)
    self.num = x
    self.den = y
  def __str__(self):
    return '{}/{}'.format(self.num, self.den)
  def __add__(self, other):
    new_num = (self.num * other.den) + (other.num * self.den)
    new_den = self.den * other.den
    return fraction(new_num, new_den)
  def __sub__(self, other):
    new_num = (self.num * other.den) - (other.num * self.den)
    new_den = self.den * other.den
    return fraction(new_num, new_den)
  def __mul__(self, other):
    new_num = (self.num * other.num)
    new_den = self.den * other.den
    return fraction(new_num, new_den)
  def __truediv__(self, other):
    new_num = (self.num * other.den)
    new_den = self.den * other.num
    return fraction(new_num, new_den)

  def convert_to_decimal(self): #non magic method no start to __
    return self.num/self.den
```

```python
fra1=fraction(3,4)
fra2=fraction(4,2)
```

```python
print(fra1 + fra2)
print(fra1-fra2)
print(fra1*fra2)
print(fra1/fra2)
fra1.convert_to_decimal()
```

```python


class product:
  def __init__(self, mfg_date, exp_date):
    self.mfg_date = mfg_date
    self.exp_date = exp_date

  def days_left(self):
    time_diff = self.exp_date - self.mfg_date
    days = time_diff.days

    years = days // 365
    remaining_days_after_years = days % 365
    months = remaining_days_after_years // 30 # Approximate months
    remaining_days = remaining_days_after_years % 30 # Remaining days

    return years, months, remaining_days # Return calculated values

# Move object creation and printing outside the class definition
mfg = datetime.date(2024, 1, 1)
exp = datetime.date(2026, 6, 15)

p = product(mfg, exp)
y, m, d = p.days_left()
print(f"years:{y}, months:{m}, days:{d}") # Corrected f-string syntax
```

```python
class phone:
  def __init__(self, brand, camera, price):
    self.brand=brand
    self.camera=camera
    self.price=price
  def buy(self):
    print("buying a phone")

class smartphone(phone):
  def buy(self):
    print("buying a smartphone")
    super().buy()
#super keyword ka use kar ke parent ka method access kar sakte hai

s=smartphone('apple',13,100000)
s.buy()
```

```python
class phone:
  def __init__(self, brand, camera, price):
    print("inside phone constructor")
    self.brand=brand
    self.camera=camera
    self.price=price
  def buy(self):
    print("buying a phone")

class smartphone(phone):
  def __init__(self,brand,camera,price,os,ram):
    print("inside smartphone constructor")
    super().__init__(brand,camera,price)
    self.os=os
    self.ram=ram
    print("inside smartphone constructor")
s=smartphone('apple',13,100000,'android',8) # is eg me pata chal raha hai ki pahle child constructor me gaye
                                             # phir parent constructor me, fir child constructor me aye
print(s.os)
print(s.ram)

```
